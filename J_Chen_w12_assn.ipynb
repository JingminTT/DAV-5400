(XML) use urllib.request.urlretrieve to get the data from github, use objectify() to parse the web page and use getroot() to get the root node.
create a empty list name "data" and create a for loop to file the data of each books, beacuse the order is different than previous one,so I use table2[['Title','Authors','Language','Pages','ISBN']] to reorder it.

```python
# load the objectify() function from the lxml library
from lxml import objectify
# request data from github and store in path and headers
path, headers = urllib.request.urlretrieve('https://raw.githubusercontent.com/TTZUO/DAV-5400/master/books.xml')
# objectify() is then used to parse the web page
parsed = objectify.parse(open(path))
# now get a reference to the root node of the XML file
root = parsed.getroot()
root
```

```python
# define an empty list that will be used to store the parsed data
data = []
# root.book is a generator that we use to extract each <book>
# element from the XML data
for i in root.book:
     # for each record, create a dict of tag names 
    el_data = {}
    for child in i.getchildren():
        el_data[child.tag] = child.pyval
    data.append(el_data)
#print data
data
```

```python
# check the results
table2 = pd.DataFrame(data)
table2
```

```python
# reorder table1
order_table2=table2[['Title','Authors','Language','Pages','ISBN']]
# print result
order_table2
```
